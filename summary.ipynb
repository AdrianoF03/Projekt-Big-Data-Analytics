```python
import pandas as pd

def load_data(file_path):
    """
    Load the dataset from the given file path.
    """
    try:
        data = pd.read_csv(file_path)
        return data
    except Exception as e:
        print(f"Error loading data: {e}")
        return None

def generate_summary_statistics(data):
    """
    Generate summary statistics for the given DataFrame.
    """
    if data is not None:
        print("\nSummary Statistics:")
        print(data.describe(include='all'))

        print("\nMissing Values:")
        print(data.isnull().sum())

        print("\nData Types:")
        print(data.dtypes)
    else:
        print("No data available for generating statistics.")

if __name__ == "__main__":
    file_path = "C:\\Users\\STENFEN\\OneDrive - Voith Group of Companies\\01 Studium und Arbeit\\01_Theoriephase\\06 Semester\\ProjektBigData\\Projekt-Big-Data-Analytics\\telefonkunden.csv"
    
    # Load the data
    df = load_data(file_path)

    # Generate summary statistics
    generate_summary_statistics(df)
```
